```python
import pandas as pd

DATA_PATH = "/content/drive/MyDrive/zillow/data"

train2016 = pd.read_csv(f"{DATA_PATH}/train_2016_v2.csv")
train2017 = pd.read_csv(f"{DATA_PATH}/train_2017.csv")
prop2016  = pd.read_csv(f"{DATA_PATH}/properties_2016.csv")
prop2017  = pd.read_csv(f"{DATA_PATH}/properties_2017.csv")
```

```python
# 2016: TRAIN
df2016 = train2016.merge(
    prop2016,
    on="parcelid",
    how="left"
)

# 2017: TEST
df2017 = train2017.merge(
    prop2017,
    on="parcelid",
    how="left"
)

print(df2016.shape)
print(df2017.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

TARGET = "logerror"
DROP_COLS = ["parcelid", "transactiondate"]

def preprocess(df):
    df = df.copy()

    y = df[TARGET].values
    X = df.drop(columns=[TARGET] + [c for c in DROP_COLS if c in df.columns])

    # SayÄ±sal kolonlar â†’ median
    num_cols = X.select_dtypes(include=["int64", "float64"]).columns
    X[num_cols] = X[num_cols].fillna(X[num_cols].median())

    # Kategorik kolonlar â†’ "None" + LabelEncoder
    cat_cols = X.select_dtypes(include=["object"]).columns
    for col in cat_cols:
        X[col] = X[col].fillna("None").astype(str)
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col])

    return X, y
```

```python
X_train, y_train = preprocess(df2016)
X_test,  y_test  = preprocess(df2017)

print(X_train.shape, X_test.shape)
```

```python
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error

baseline = LGBMRegressor(
    random_state=42,
    n_jobs=-1
)

baseline.fit(X_train, y_train)   # ðŸ”¥ EÄžÄ°TÄ°M

pred_base = baseline.predict(X_test)
mae_base = mean_absolute_error(y_test, pred_base)

mae_base
```

```python

# =========================
# KURULUMLAR
# =========================
!pip install -q lightgbm xgboost catboost

# =========================
# IMPORTLAR
# =========================
import time
import numpy as np
import pandas as pd

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor

# =========================
# METRÄ°K FONKSÄ°YONU
# =========================
def evaluate_model(name, model, X_train, y_train, X_test, y_test):
    start = time.time()
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    end = time.time()

    n = len(y_test)
    p = X_train.shape[1]

    return {
        "Model": name,
        "MAE": mean_absolute_error(y_test, preds),
        "RMSE": np.sqrt(mean_squared_error(y_test, preds)),
        "Median AE": np.median(np.abs(y_test - preds)),
        "MAPE (%)": np.mean(np.abs((y_test - preds) / (y_test + 1e-9))) * 100,
        "R2": r2_score(y_test, preds),
        "Adjusted R2": 1 - (1 - r2_score(y_test, preds)) * (n - 1) / (n - p - 1),
        "Train Time (s)": end - start
    }

# =========================
# MODELLER (GPU + DEFAULT)
# =========================
models = [
    (
        "LightGBM (DEFAULT)",
        LGBMRegressor(
            device="gpu",
            random_state=42
        )
    ),
    (
        "XGBoost (DEFAULT)",
        XGBRegressor(
        tree_method="hist",
        random_state=42
      )
    ),
    (
        "CatBoost (DEFAULT)",
        CatBoostRegressor(
            task_type="GPU",
            verbose=0,
            random_state=42
        )
    )
]

# =========================
# EÄžÄ°TÄ°M + TEST
# =========================
results = []

for name, model in models:
    print(f"\n=== {name} ===")
    res = evaluate_model(
        name,
        model,
        X_train, y_train,
        X_test, y_test
    )
    results.append(res)

# =========================
# SONUÃ‡ TABLOSU
# =========================
results_boost_default_df = (
    pd.DataFrame(results)
    .sort_values("MAE")
    .reset_index(drop=True)
)

results_boost_default_df
```

```python
# =========================================================
# 0. IMPORTS
# =========================================================
import time
import numpy as np
import pandas as pd

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor

# =========================================================
# 1. METRIK FONKSIYONU
# =========================================================
def compute_metrics(y_true, preds, X_train):
    mae = mean_absolute_error(y_true, preds)
    rmse = np.sqrt(mean_squared_error(y_true, preds))
    median_ae = np.median(np.abs(y_true - preds))
    mape = np.mean(np.abs((y_true - preds) / (y_true + 1e-9))) * 100
    r2 = r2_score(y_true, preds)

    n = len(y_true)
    p = X_train.shape[1]
    adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

    return mae, rmse, median_ae, mape, r2, adj_r2

# =========================================================
# 2. DEFAULT BOOST MODELLER
# =========================================================
results_default = []

default_models = {
    "LightGBM (DEFAULT)": LGBMRegressor(
        device="gpu",
        random_state=42
    ),
    "XGBoost (DEFAULT)": XGBRegressor(
        objective="reg:squarederror",
        tree_method="hist",
        random_state=42
    ),
    "CatBoost (DEFAULT)": CatBoostRegressor(
        loss_function="MAE",
        task_type="GPU",
        devices="0",
        verbose=False,
        random_seed=42
    )
}

for name, model in default_models.items():
    print(f"\n=== {name} ===")
    start = time.time()
    model.fit(X_train, y_train)
    end = time.time()

    preds = model.predict(X_test)
    mae, rmse, median_ae, mape, r2, adj_r2 = compute_metrics(y_test, preds, X_train)

    results_default.append({
        "Model": name,
        "MAE": mae,
        "RMSE": rmse,
        "Median AE": median_ae,
        "MAPE (%)": mape,
        "R2": r2,
        "Adjusted R2": adj_r2,
        "Train Time (s)": end - start
    })

df_default = pd.DataFrame(results_default).sort_values("MAE")
df_default

# =========================================================
# 3. TUNED BOOST MODELLER (GPU + KÃœÃ‡ÃœK GRID)
# =========================================================
results_tuned = []

tuned_models = {
    "LightGBM (TUNED)": {
        "model": LGBMRegressor(device="gpu", random_state=42),
        "params": {
            "n_estimators": [300, 600],
            "learning_rate": [0.03, 0.05],
            "num_leaves": [31, 63],
            "max_depth": [-1, 10]
        }
    },
    "XGBoost (TUNED)": {
        "model": XGBRegressor(
            objective="reg:squarederror",
            tree_method="hist",
            random_state=42
        ),
        "params": {
            "n_estimators": [300, 600],
            "learning_rate": [0.05],
            "max_depth": [6, 8]
        }
    },
    "CatBoost (TUNED)": {
        "model": CatBoostRegressor(
            loss_function="MAE",
            task_type="GPU",
            devices="0",
            verbose=False,
            random_seed=42
        ),
        "params": {
            "iterations": [300, 600],
            "learning_rate": [0.03, 0.05],
            "depth": [6, 8]
        }
    }
}

for name, cfg in tuned_models.items():
    print(f"\n=== {name} ===")

    search = GridSearchCV(
        cfg["model"],
        cfg["params"],
        scoring="neg_mean_absolute_error",
        cv=3,
        n_jobs=-1
    )

    start = time.time()
    search.fit(X_train, y_train)
    end = time.time()

    best_model = search.best_estimator_
    print("Best params:", search.best_params_)

    preds = best_model.predict(X_test)
    mae, rmse, median_ae, mape, r2, adj_r2 = compute_metrics(y_test, preds, X_train)

    results_tuned.append({
        "Model": name,
        "MAE": mae,
        "RMSE": rmse,
        "Median AE": median_ae,
        "MAPE (%)": mape,
        "R2": r2,
        "Adjusted R2": adj_r2,
        "Train Time (s)": end - start
    })

df_tuned = pd.DataFrame(results_tuned).sort_values("MAE")
df_tuned

# =========================================================
# 4. FINAL KARÅžILAÅžTIRMA
# =========================================================
final_results = pd.concat([df_default, df_tuned]).sort_values("MAE")
final_results
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error

def bootstrap_mae_diff(
    y_true,
    pred_a,
    pred_b,
    n_bootstrap=10000,
    random_state=42
):
    """
    pred_a: Default model predictions
    pred_b: Tuned model predictions
    Test edilen fark: MAE(pred_b) - MAE(pred_a)
    (Negatif deÄŸer = tuned daha iyi)
    """
    rng = np.random.RandomState(random_state)
    n = len(y_true)

    diffs = []

    for _ in range(n_bootstrap):
        idx = rng.choice(n, n, replace=True)
        mae_a = mean_absolute_error(y_true[idx], pred_a[idx])
        mae_b = mean_absolute_error(y_true[idx], pred_b[idx])
        diffs.append(mae_b - mae_a)

    diffs = np.array(diffs)

    ci_low, ci_high = np.percentile(diffs, [2.5, 97.5])
    mean_diff = diffs.mean()

    p_value = np.mean(diffs >= 0)  # tuned'Ä±n daha kÃ¶tÃ¼ olma olasÄ±lÄ±ÄŸÄ±

    return {
        "Mean MAE diff (tuned - default)": mean_diff,
        "95% CI lower": ci_low,
        "95% CI upper": ci_high,
        "P(tuned >= default)": p_value
    }
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error

def bootstrap_ci_mae(y_true, y_pred, n_bootstrap=10000, alpha=0.05):
    rng = np.random.RandomState(42)
    n = len(y_true)
    maes = []

    for _ in range(n_bootstrap):
        idx = rng.choice(n, n, replace=True)
        maes.append(mean_absolute_error(y_true[idx], y_pred[idx]))

    lower = np.percentile(maes, 100 * alpha / 2)
    upper = np.percentile(maes, 100 * (1 - alpha / 2))

    return {
        "MAE_mean": np.mean(maes),
        "CI_lower": lower,
        "CI_upper": upper
    }
```
